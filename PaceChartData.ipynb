```python
import pandas as pd
```

```python
times_3k = pd.read_csv(r"C:\Users\Jason\Downloads\DataScience\3k_Times.csv")
times_5k = pd.read_csv(r"C:\Users\Jason\Downloads\DataScience\5k_Times.csv")
times_10k = pd.read_csv(r"C:\Users\Jason\Downloads\DataScience\10k_Times.csv")
times_hm = pd.read_csv(r"C:\Users\Jason\Downloads\DataScience\hm_Times.csv")
times_marathon = pd.read_csv(r"C:\Users\Jason\Downloads\DataScience\mara_Times.csv", encoding='latin1')
```

```python
def format_to_time(sec):
    tot = round(sec)
    minutes = tot // 60
    remaining_seconds = tot % 60
    return f"{minutes:02}:{remaining_seconds:02}"

def format_to_sec(time_str):
    minutes, seconds = time_str.split(":")
    return int(minutes) * 60 + float(seconds)
```

```python
comp = times_3k.merge(times_5k, on='Name', how='left').merge(times_10k, on='Name', how='left').merge(times_hm, on='Name', how='left')
comp.dropna(inplace=True)
comp['3k_Time'] = comp['3k_Time'].apply(format_to_sec)
comp['5k_Time'] = comp['5k_Time'].apply(format_to_sec)
comp['10k_Time'] = comp['10k_Time'].apply(format_to_sec)
comp['hm_Time'] = pd.to_timedelta(comp['hm_Time']).dt.total_seconds().astype(int)
comp['3k_Pace'] = (comp['3k_Time'] / 1.86).round(2)
comp['5k_Pace'] = (comp['5k_Time'] / 3.1).round(2)
comp['10k_Pace'] = (comp['10k_Time'] / 6.2).round(2)
comp['hm_Pace'] = (comp['hm_Time'] / 13.1).round(2)
comp['5k_perc'] = (comp['3k_Pace'] / comp['5k_Pace']).round(4)
comp['10k_perc'] = (comp['3k_Pace'] / comp['10k_Pace']).round(4)
comp['hm_perc'] = (comp['3k_Pace'] / comp['hm_Pace']).round(4)
comp_m = comp.merge(times_marathon, on='Name', how='left').dropna()
comp_m['Marathon'] = pd.to_timedelta(comp_m['Marathon']).dt.total_seconds().astype(int)
comp_m['Marathon_Pace'] = (comp_m['Marathon'] / 26.2).round(2)
comp_m['Marathon_perc'] = (comp_m['3k_Pace'] / comp_m['Marathon_Pace']).round(4)
comp_m
```
