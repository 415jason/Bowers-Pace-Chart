```python

import numpy as np
import pandas as pd
```

```python
temps = pd.read_csv(r"C:\Users\jason\Downloads\TempNormals.csv")
humid = pd.read_csv(r"C:\Users\jason\Downloads\HumidityData_Formatted.csv")
```

```python
#Dropping columns for afternoon data, as the majority of races are in the morning
humid.drop(columns=['Jan_a', 'Feb_a', 'Mar_a', 'Apr_a', 'May_a', 'Jun_a', 'Jul_a', 'Aug_a', 'Sep_a', 'Oct_a', 'Nov_a', 'Dec_a', 'Ann_a'], inplace=True)
humid['Start_Date'] = humid['Date'].str[0:4]
humid['End_Date'] = humid['Date'].str[5:9]
humid.drop(columns=['Date'], inplace=True)
humid['Median_Date'] = (humid['End_Date'].astype(int) + humid['Start_Date'].astype(int)) / 2
#Converting relative humidity values based on 2020 humidity trends
humid['Jan_m'] = humid['Jan_m'] - (0.05 * (2020 - humid['Median_Date']))
humid['Feb_m'] = humid['Feb_m'] - (0.05 * (2020 - humid['Median_Date']))
humid['Mar_m'] = humid['Mar_m'] - (0.05 * (2020 - humid['Median_Date']))
humid['Apr_m'] = humid['Apr_m'] - (0.05 * (2020 - humid['Median_Date']))
humid['May_m'] = humid['May_m'] - (0.05 * (2020 - humid['Median_Date']))
humid['Jun_m'] = humid['Jun_m'] - (0.05 * (2020 - humid['Median_Date']))
humid['Jul_m'] = humid['Jul_m'] - (0.05 * (2020 - humid['Median_Date']))
humid['Aug_m'] = humid['Aug_m'] - (0.05 * (2020 - humid['Median_Date']))
humid['Sep_m'] = humid['Sep_m'] - (0.05 * (2020 - humid['Median_Date']))
humid['Oct_m'] = humid['Oct_m'] - (0.05 * (2020 - humid['Median_Date']))
humid['Nov_m'] = humid['Nov_m'] - (0.05 * (2020 - humid['Median_Date']))
humid['Dec_m'] = humid['Dec_m'] - (0.05 * (2020 - humid['Median_Date']))
#Converting temperature values based on 2020 temperature trends
temps['January'] = temps['January'] + (14.5 * .025)
temps['February'] = temps['February'] + (14.5 * .025)
temps['March'] = temps['March'] + (14.5 * .025)
temps['April'] = temps['April'] + (14.5 * .025)
temps['May'] = temps['May'] + (14.5 * .025)
temps['June'] = temps['June'] + (14.5 * .025)
temps['July'] = temps['July'] + (14.5 * .025)
temps['August'] = temps['August'] + (14.5 * .025)
temps['September'] = temps['September'] + (14.5 * .025)
temps['October'] = temps['October'] + (14.5 * .025)
temps['November'] = temps['November'] + (14.5 * .025)
temps['December'] = temps['December'] + (14.5 * .025)
humid['Lon'] = humid['Lon'] * -1
#Creating columns for tempeture so that they can be added into the dataframe in the next cell
humid['Jan_Temp_avg'] = np.nan
humid['Feb_Temp_avg'] = np.nan
humid['Mar_Temp_avg'] = np.nan
humid['Apr_Temp_avg'] = np.nan
humid['May_Temp_avg'] = np.nan
humid['Jun_Temp_avg'] = np.nan
humid['Jul_Temp_avg'] = np.nan
humid['Aug_Temp_avg'] = np.nan
humid['Sep_Temp_avg'] = np.nan
humid['Oct_Temp_avg'] = np.nan
humid['Nov_Temp_avg'] = np.nan
humid['Dec_Temp_avg'] = np.nan
humid.head(10)
```

```python
tempJanList = []
tempFebList = []
tempMarList = []
tempAprList = []
tempMayList = []
tempJunList = []
tempJulList = []
tempAugList = []
tempSepList = []
tempOctList = []
tempNovList = []
tempDecList = []
i = 0
j = 0
while i < humid['City'].count():
    while j < temps['Lat'].count():
        #Process to make iteration more efficient
        if (float(j < (temps['Lat'].count() - 25000)) and (float(temps.iloc[j+25000, 0]) < float(humid.iloc[i, 1]-.16))):
            j += 24999
        elif (float(j < (temps['Lat'].count() - 10000)) and (float(temps.iloc[j+10000, 0]) < float(humid.iloc[i, 1]-.16))):
            j += 9999
        elif (float(j < (temps['Lat'].count() - 4000)) and ((float(temps.iloc[j+4000, 0]) < float(humid.iloc[i, 1]-.16)))):
            j += 3999
        if ((float(humid.iloc[i, 2]) + .15) > float(temps.iloc[j, 1])) and (float(temps.iloc[j, 1]) > (float(humid.iloc[i, 2]) - .15)) and ((float(humid.iloc[i, 1]) + .15) > float(temps.iloc[j, 0])) and (float(temps.iloc[j, 0]) > (float(humid.iloc[i, 1]) - .15)):
            #If the coordinates in the tempature data are within .15 degrees of the coordinates in the humidity data, then the tempature data is stored
            tempJanList.append(temps.loc[i, 'January'])
            tempFebList.append(temps.loc[i, 'February'])
            tempMarList.append(temps.loc[i, 'March'])
            tempAprList.append(temps.loc[i, 'April'])
            tempMayList.append(temps.loc[i, 'May'])
            tempJunList.append(temps.loc[i, 'June'])
            tempJulList.append(temps.loc[i, 'July'])
            tempAugList.append(temps.loc[i, 'August'])
            tempSepList.append(temps.loc[i, 'September'])
            tempOctList.append(temps.loc[i, 'October'])
            tempNovList.append(temps.loc[i, 'November'])
            tempDecList.append(temps.loc[i, 'December'])
        elif (float(temps.iloc[j, 0]) > float(humid.iloc[i, 1]+.16)):
            j = 1000000
        j += 1
    if len(tempFebList) > 0:
        #Takes the avaerage of all the coordinates that were matched to the city to give a single value for tempeture
        humid.loc[i, 'Jan_Temp_avg'] = (np.mean(tempJanList)).round(2)
        humid.loc[i, 'Feb_Temp_avg'] = (np.mean(tempFebList)).round(2)
        humid.loc[i, 'Mar_Temp_avg'] = (np.mean(tempMarList)).round(2)
        humid.loc[i, 'Apr_Temp_avg'] = (np.mean(tempAprList)).round(2)
        humid.loc[i, 'May_Temp_avg'] = (np.mean(tempMayList)).round(2)
        humid.loc[i, 'Jun_Temp_avg'] = (np.mean(tempJunList)).round(2)
        humid.loc[i, 'Jul_Temp_avg'] = (np.mean(tempJulList)).round(2)
        humid.loc[i, 'Aug_Temp_avg'] = (np.mean(tempAugList)).round(2)
        humid.loc[i, 'Sep_Temp_avg'] = (np.mean(tempSepList)).round(2)
        humid.loc[i, 'Oct_Temp_avg'] = (np.mean(tempOctList)).round(2)
        humid.loc[i, 'Nov_Temp_avg'] = (np.mean(tempNovList)).round(2)
        humid.loc[i, 'Dec_Temp_avg'] = (np.mean(tempDecList)).round(2)
    tempJanList.clear()
    tempFebList.clear()
    tempMarList.clear()
    tempAprList.clear()
    tempMayList.clear()
    tempJunList.clear()
    tempJulList.clear()
    tempAugList.clear()
    tempSepList.clear()
    tempOctList.clear()
    tempNovList.clear()
    tempDecList.clear()
    i += 1
    j = 0
```

```python
#Dropping the cities that did not have any matches in the temperature dataset 
if (humid.isnull().values.any()):
    humid.dropna(inplace=True)
#Coverting the relative humidity to dew point
humid['Jan_Dew'] = (humid['Jan_Temp_avg'] - ((100 - humid['Jan_m']) / 5)).round(2)
humid['Feb_Dew'] = (humid['Feb_Temp_avg'] - ((100 - humid['Feb_m']) / 5)).round(2)
humid['Mar_Dew'] = (humid['Mar_Temp_avg'] - ((100 - humid['Mar_m']) / 5)).round(2)
humid['Apr_Dew'] = (humid['Apr_Temp_avg'] - ((100 - humid['Apr_m']) / 5)).round(2)
humid['May_Dew'] = (humid['May_Temp_avg'] - ((100 - humid['May_m']) / 5)).round(2)
humid['Jun_Dew'] = (humid['Jun_Temp_avg'] - ((100 - humid['Jun_m']) / 5)).round(2)
humid['Jul_Dew'] = (humid['Jul_Temp_avg'] - ((100 - humid['Jul_m']) / 5)).round(2)
humid['Aug_Dew'] = (humid['Aug_Temp_avg'] - ((100 - humid['Aug_m']) / 5)).round(2)
humid['Sep_Dew'] = (humid['Sep_Temp_avg'] - ((100 - humid['Sep_m']) / 5)).round(2)
humid['Oct_Dew'] = (humid['Oct_Temp_avg'] - ((100 - humid['Oct_m']) / 5)).round(2)
humid['Nov_Dew'] = (humid['Nov_Temp_avg'] - ((100 - humid['Nov_m']) / 5)).round(2)
humid['Dec_Dew'] = (humid['Dec_Temp_avg'] - ((100 - humid['Dec_m']) / 5)).round(2)
humid.to_csv(r"C:\Users\Jason\Downloads\Humidity_Final.csv", index=False)
humid.head(10)
```
