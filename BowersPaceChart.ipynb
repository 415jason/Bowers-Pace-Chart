```python
df['8:00 TT Pace'] = (round(df['8:00 TT Pace'])).astype(int)
df['5k Pace'] = (round(df['5k Pace'])).astype(int)
df['10k Pace'] = (round(df['10k Pace'])).astype(int)
df['hm Pace'] = (round(df['hm Pace'])).astype(int)
df['5k Lower'] = (round(df['5k Pace'] * 1.00402898)).astype(int)
df['5k Upper'] = (round(df['5k Pace'] * 1.047778)).astype(int)
df['10k Lower'] = (round(df['10k Pace'] * 1.00402898)).astype(int)
df['10k Upper'] = (round(df['10k Pace'] * 1.047778)).astype(int)
df['hm Lower'] = (round(df['hm Pace'] * 1.00402898)).astype(int)
df['hm Upper'] = (round(df['hm Pace'] * 1.047778)).astype(int)
df['com Lower'] = (round(df['8:00 TT Pace'] / .76)).astype(int)
df['com Upper'] = (round(df['8:00 TT Pace'] / .71)).astype(int)
df['ve Lower'] = (round(df['8:00 TT Pace'] / .7)).astype(int)
df['ve Upper'] = (round(df['8:00 TT Pace'] / .65)).astype(int)
df
```

```python
time_df = df
time_df.drop('perc5k', axis=1, inplace=True)
time_df.drop('perc10k', axis=1, inplace=True)
time_df.drop('perc_hm', axis=1, inplace=True)
time_df['8:00 TT Pace'] = df['8:00 TT Pace'].apply(lambda x: f"{x // 60:02}:{x % 60:02}")
time_df['5k Pace'] = df['5k Pace'].apply(lambda x: f"{x // 60:02}:{x % 60:02}")
time_df['10k Pace'] = df['10k Pace'].apply(lambda x: f"{x // 60:02}:{x % 60:02}")
time_df['hm Pace'] = df['hm Pace'].apply(lambda x: f"{x // 60:02}:{x % 60:02}")
time_df['5k Upper'] = df['5k Upper'].apply(lambda x: f"{x // 60:02}:{x % 60:02}")
time_df['5k Lower'] = df['5k Lower'].apply(lambda x: f"{x // 60:02}:{x % 60:02}")
time_df['10k Upper'] = df['10k Upper'].apply(lambda x: f"{x // 60:02}:{x % 60:02}")
time_df['10k Lower'] = df['10k Lower'].apply(lambda x: f"{x // 60:02}:{x % 60:02}")
time_df['hm Upper'] = df['hm Upper'].apply(lambda x: f"{x // 60:02}:{x % 60:02}")
time_df['hm Lower'] = df['hm Lower'].apply(lambda x: f"{x // 60:02}:{x % 60:02}")
time_df['com Upper'] = df['com Upper'].apply(lambda x: f"{x // 60:02}:{x % 60:02}")
time_df['com Lower'] = df['com Lower'].apply(lambda x: f"{x // 60:02}:{x % 60:02}")
time_df['ve Upper'] = df['ve Upper'].apply(lambda x: f"{x // 60:02}:{x % 60:02}")
time_df['ve Lower'] = df['ve Lower'].apply(lambda x: f"{x // 60:02}:{x % 60:02}")
time_df
```

```python
final_df = pd.DataFrame()
final_df['8:00 TT Pace'] = time_df['8:00 TT Pace'].copy()
final_df['5k Pace Range'] = time_df['5k Lower'].astype(str) + " - " + time_df['5k Upper'].astype(str)
final_df['10k Pace Range'] = time_df['10k Lower'].astype(str) + " - " + time_df['10k Upper'].astype(str)
final_df['Half Marathon Pace Range'] = time_df['hm Lower'].astype(str) + " - " + time_df['hm Upper'].astype(str)
final_df['Comfortable Pace Range'] = time_df['com Lower'].astype(str) + " - " + time_df['com Upper'].astype(str)
final_df['Very Easy Pace Range'] = time_df['ve Lower'].astype(str) + " - " + time_df['ve Upper'].astype(str)
final_df
```
